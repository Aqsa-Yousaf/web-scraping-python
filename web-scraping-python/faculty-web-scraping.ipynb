```python
import requests
from bs4 import BeautifulSoup
import csv
import os

#  Request the page
url = 'https://cecos.edu.pk/computer-science/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Create a CSV file to save faculty data
csv_file = open('faculty_data.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Name', 'Designation', 'Department', 'Email', 'Contact'])

#  Scrape faculty information
faculty_section = soup.find_all('div', class_='elementor-widget-text-editor')
for faculty in faculty_section:
    try:
        name = faculty.find('h6').text.strip()
        designation = faculty.find('p').text.strip().split("–")[0].strip()
        department = "Computer Science"  
        email = "N/A"  # Email not available 
        contact = "N/A"  # Contact info not available 
        
        # Save the data into the CSV file
        csv_writer.writerow([name, designation, department, email, contact])
    except AttributeError:
        continue


#  Create a text file with University information
with open('university_info.txt', 'w') as f:
    f.write(f"University Name: CECOS University\n")
    f.write(f"Faculty Page URL: {url}")

# Close the CSV file
csv_file.close()

print("Data scraping complete.")
```
